```python
def get_blog_search_url(keyword, start_date, end_date, page):
    base_url = "https://section.blog.naver.com/Search/Post.naver?"
    return f"{base_url}pageNo={page}&rangeType=WEEK&orderBy=sim&startDate={start_date}&endDate={end_date}&keyword={keyword}"

def get_one_page_blog_urls(driver):
    soup = BeautifulSoup(driver.page_source, "html.parser")
    return [a["href"] for a in soup.select("div > a.desc_inner")]

def crawl_blog_urls(keywords, start_date, end_date, filename):
    driver = webdriver.Chrome()
    total_blog_urls = []
    keyword_counts = {}
    for keyword in keywords:
        keyword_counts[keyword] = 0
        print(f"'{keyword}' ÌÇ§ÏõåÎìú ÌÅ¨Î°§ÎßÅ Ï§ë")
        for page in range(1, 1100 // 7 + 2):  # 1ÌéòÏù¥ÏßÄÎãπ ÏïΩ 7Í∞ú Í∏∞Ï§Ä, 1100Í∞ú ÎèÑÎã¨ Ïãú Ï¢ÖÎ£å
            if keyword_counts[keyword] >= 1100:
                break
            driver.get(get_blog_search_url(keyword, start_date, end_date, page))
            time.sleep(2)
            one_page_blog_urls = get_one_page_blog_urls(driver)
            remaining = 1100 - keyword_counts[keyword]
            total_blog_urls.extend(one_page_blog_urls[:remaining])
            keyword_counts[keyword] += len(one_page_blog_urls[:remaining])
        print(f"'{keyword}' ÌÅ¨Î°§ÎßÅ ÏôÑÎ£å. Ï¥ù {keyword_counts[keyword]}Í∞ú\n")
    unique_blog_urls = list(set(total_blog_urls))
    df = pd.DataFrame({"link": unique_blog_urls})
    df.to_csv(filename, index=False)
    driver.quit()
```

```python
keywords = ["ÏùòÎ•òÍ¥ÄÎ¶¨Í∏∞ -lgÏä§ÌÉÄÏùºÎü¨ -ÏóêÏñ¥ÎìúÎ†àÏÑú", "lgÏä§ÌÉÄÏùºÎü¨ -ai", "ÏóêÏñ¥ÎìúÎ†àÏÑú -ai", "ai lgÏä§ÌÉÄÏùºÎü¨", "ai ÏóêÏñ¥ÎìúÎ†àÏÑú"]
crawl_blog_urls(keywords, "2022-01-01", "2022-12-31", "ÏùòÎ•òÍ¥ÄÎ¶¨Í∏∞_2022.csv")
```

```python
# Í≤ÄÏÉâ URL ÏÉùÏÑ± Ìï®Ïàò
def get_blog_search_url(keyword, start_date, end_date, page):
    base_url = "https://section.blog.naver.com/Search/Post.naver?"
    return f"{base_url}pageNo={page}&rangeType=WEEK&orderBy=sim&startDate={start_date}&endDate={end_date}&keyword={keyword}"

# Ìïú ÌéòÏù¥ÏßÄÏóêÏÑú Î∏îÎ°úÍ∑∏ URL Ï∂îÏ∂ú Ìï®Ïàò
def get_one_page_blog_urls(driver):
    soup = BeautifulSoup(driver.page_source, "html.parser")
    return [a["href"] for a in soup.select("div > a.desc_inner")]

# ÏµúÎåÄ ÌéòÏù¥ÏßÄ Ïàò Í∞ÄÏ†∏Ïò§Îäî Ìï®Ïàò
def get_max_page(driver):
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    
    try:
        # ÌòÑÏû¨ ÌéòÏù¥ÏßÄÏóêÏÑú ÌëúÏãúÎêú ÌéòÏù¥ÏßÄ Î≤àÌò∏ Ï∞æÍ∏∞
        page_elements = soup.select("#content > section > div.pagination > span > a")
        page_numbers = [int(a.text) for a in page_elements if a.text.isdigit()]
        
        # 10ÌéòÏù¥ÏßÄ Ïù¥ÏÉÅÏù∏ÏßÄ ÌôïÏù∏
        while True:
            try:
                # "Îã§Ïùå" Î≤ÑÌäº ÌÅ¥Î¶≠, Ï∂îÍ∞Ä ÌéòÏù¥ÏßÄ ÌôïÏù∏
                next_button = driver.find_element(By.CLASS_NAME, "button_next")
                
                if "disabled" in next_button.get_attribute("class"):
                    break  # "Îã§Ïùå" Î≤ÑÌäºÏù¥ ÎπÑÌôúÏÑ±ÌôîÎêòÎ©¥ Ï¢ÖÎ£å

                next_button.click()  # "Îã§Ïùå" Î≤ÑÌäº ÌÅ¥Î¶≠
                time.sleep(2)
                
                # "Îã§Ïùå" ÌÅ¥Î¶≠ ÌõÑ Îã§Ïãú ÌéòÏù¥ÏßÄ Î≤àÌò∏ ÏóÖÎç∞Ïù¥Ìä∏
                soup = BeautifulSoup(driver.page_source, "html.parser")
                new_page_elements = soup.select("#content > section > div.pagination > span > a")
                new_page_numbers = [int(a.text) for a in new_page_elements if a.text.isdigit()]
                
                # ÏÉàÎ°úÏö¥ ÌéòÏù¥ÏßÄ Î≤àÌò∏ Ï∂îÍ∞Ä
                page_numbers.extend(new_page_numbers)
                page_numbers = list(set(page_numbers))  # Ï§ëÎ≥µ Ï†úÍ±∞
                
            except Exception:
                break  # "Îã§Ïùå Î≤ÑÌäº"Ïù¥ ÏóÜÏúºÎ©¥ Ï¢ÖÎ£å
        
        max_page = max(page_numbers) if page_numbers else 1  # ÏµúÎåÄ ÌéòÏù¥ÏßÄ Î≤àÌò∏ Î∞òÌôò
        return max_page
    except Exception as e:
        print(f"ÏµúÎåÄ ÌéòÏù¥ÏßÄ Ï∞æÍ∏∞ Ïã§Ìå®: {e}")
        return 1  # Ïò§Î•ò Î∞úÏÉù Ïãú Í∏∞Î≥∏Í∞í 1ÌéòÏù¥ÏßÄ Î∞òÌôò

# ÌÅ¨Î°§ÎßÅ Ïã§Ìñâ
keywords = ["ÏùòÎ•òÍ¥ÄÎ¶¨Í∏∞ -lgÏä§ÌÉÄÏùºÎü¨ -ÏóêÏñ¥ÎìúÎ†àÏÑú", "lgÏä§ÌÉÄÏùºÎü¨ -ai", "ÏóêÏñ¥ÎìúÎ†àÏÑú -ai", 'ai lgÏä§ÌÉÄÏùºÎü¨', 'ai ÏóêÏñ¥ÎìúÎ†àÏÑú']
start_date = "2020-01-01"
end_date = "2021-12-31"

total_blog_urls = []  # Ï†ÑÏ≤¥ Î∏îÎ°úÍ∑∏ URL Ï†ÄÏû•
keyword_counts = {}   # ÌÇ§ÏõåÎìúÎ≥Ñ ÌÅ¨Î°§ÎßÅ Í∞úÏàò Ï†ÄÏû•

for keyword in keywords:
    keyword_counts[keyword] = 0  # Ï¥àÍ∏∞ Í∞úÏàò ÏÑ§Ï†ï
    driver.get(get_blog_search_url(keyword, start_date, end_date, 1))
    time.sleep(3)

    max_page = get_max_page(driver)  # ÏµúÎåÄ ÌéòÏù¥ÏßÄ Ïàò Í∞ÄÏ†∏Ïò§Í∏∞
    print(f"'{keyword}' ÌÇ§ÏõåÎìú - ÏµúÎåÄ {max_page} ÌéòÏù¥ÏßÄÍπåÏßÄ ÌÅ¨Î°§ÎßÅ Ï§ë")

    for page in range(1, max_page + 1):
        driver.get(get_blog_search_url(keyword, start_date, end_date, page))
        time.sleep(2)

        # Î∏îÎ°úÍ∑∏ URL ÌÅ¨Î°§ÎßÅ
        one_page_blog_urls = get_one_page_blog_urls(driver)
        total_blog_urls.extend(one_page_blog_urls)
        keyword_counts[keyword] += len(one_page_blog_urls)

    print(f"'{keyword}' ÌÅ¨Î°§ÎßÅ ÏôÑÎ£å, Ï¥ù {keyword_counts[keyword]}Í∞ú\n")

df = pd.DataFrame({"blog_url": total_blog_urls})
```

```python
def get_content(driver):
    time.sleep(1)  # ÌéòÏù¥ÏßÄ Î°úÎî© ÎåÄÍ∏∞

    # ÌåùÏóÖ Í∞êÏßÄ Î∞è ÏûêÎèô Îã´Í∏∞
    try:
        alert = Alert(driver)
        alert.dismiss()  
        time.sleep(1)
    except Exception:
        pass  # ÌåùÏóÖÏù¥ ÏóÜÏúºÎ©¥ Í∑∏ÎÉ• ÏßÑÌñâ

    soup = BeautifulSoup(driver.page_source, "html.parser")

    try:
        iframe = driver.find_element(By.CSS_SELECTOR, "iframe#mainFrame")
        driver.switch_to.frame(iframe)
        time.sleep(1)
        soup = BeautifulSoup(driver.page_source, "html.parser")
    except NoSuchElementException:
        pass

    try:
        # ÎÇ†Ïßú
        date = soup.select_one('span.se_publishDate, span.post-date').get_text().strip()
    except:
        date = ''

    try:
        # Ï†úÎ™©
        title = soup.select_one('div.se-title-text, h3#title_area, h3.post-title').get_text().strip()
    except:
        title = ''

    try:
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.se-main-container")))
        print("Î≥∏Î¨∏ Î°úÎìú ÏôÑÎ£å")
    except:
        print("Î≥∏Î¨∏ Î°úÎìú Ïã§Ìå® (Timeout)")

    try:
        # ÎÇ¥Ïö©
        content = soup.select_one('div.se-main-container').get_text()
        
    except:
        content = ''

    # URL
    link = driver.current_url
    return [title, content, date, link]
```

```python
urls = pd.read_csv("ÏùòÎ•òÍ¥ÄÎ¶¨Í∏∞url_2024.csv")
naver_urls = urls['link'].tolist()
```

```python
results = []
count = 0

for url in naver_urls:
    driver.get(url)
    time.sleep(1)

    blog_data = get_content(driver)
    results.append(blog_data)
    count += 1

    if count % 100 == 0:
        print(f"{count}Í∞ú ÌÅ¨Î°§ÎßÅ ÏôÑÎ£å")
    
    
# df ÌòïÌÉúÎ°ú Ï†ÄÏû•
df = pd.DataFrame(results)
df["platform"] = "blog"

df.columns = ['title', 'content', 'date', 'link', 'platform']
df.to_csv('ÏùòÎ•òÍ¥ÄÎ¶¨Í∏∞_2024.csv')
```
